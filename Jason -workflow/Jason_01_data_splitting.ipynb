# Data Splitting

```python
import pandas as pd
import os
```

```python
consumer_df = pd.read_parquet("/Users/jasonc/Desktop/DSC_291/cashflow/consumer_data.parquet")
```

```python
transaction_df = pd.read_parquet("/Users/jasonc/Desktop/DSC_291/cashflow/transactions.parquet")
```

```python
consumer_df.head()
```

```python
transaction_df.head()
```

```python
# Get a number of unique consumers first:
unique_consumers = transaction_df['masked_consumer_id'].unique()
print(f"Total unique consumers: {len(unique_consumers)}")
```

```python
# Dictionary: key = consumer ID, value = their transactions (as DataFrame)
consumer_transactions = {
    consumer_id: group.reset_index(drop=True)
    for consumer_id, group in transaction_df.groupby("masked_consumer_id")
}
```

```python
# Save separated transactions:
output_dir = "/Users/jasonc/Desktop/DSC_291/split_transactions"
os.makedirs(output_dir, exist_ok=True)

for consumer_id, df in consumer_transactions.items():
    df.to_parquet(f"{output_dir}/{consumer_id}.parquet", index=False)
```
